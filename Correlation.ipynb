```python
# Подключаем необходимые для анализа библиотеки

from IPython.display import display
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
from scipy.stats import norm
import math as ms
import numpy as np
```

```python
def normal_plot(df, array):
    voters = array
    x = voters
    
    source = {"Index": ["ind"]*len(x), "voters" : x} # Индексируем сырой массив из кол-ва избирателей представленной выборки

    df = pd.DataFrame(source)

    df_mean = np.mean(df["voters"])
    print('\nСреднее значение данной выборки = ', df_mean)
    
    df_std = np.std(df["voters"])
    print('\nНормальное отклонение  данной выборки = ', df_std)
    
    # Расчёт Probability Density Function/Функция плотности вероятности (array, mu, sigma)
    # Применяется на сортированный массив
    pdf = norm.pdf(df["voters"].sort_values(), df_mean, df_std)
    
    plt.plot(df["voters"].sort_values(), pdf)
    plt.xlim([-1200,4000])  
    plt.xlabel("Кол-во избирателей", size=16)    
    plt.ylabel("Частота", size=16)                
    plt.grid(True, alpha=0.3, linestyle="--")
    plt.show()
```

```python
# Формируем Mechanical DataFrame

mech_path ="./CSV Samples/mech_df.csv"
mech_df = pd.read_csv(mech_path)
mech_df=mech_df.drop('index',axis=1)

display(mech_df)
```

```python
# Формируем отдельный DataFrame только с кол-вом избирателей для расчётов

mech_voters = mech_df['raw_voters']
print(mech_voters)
```

```python
df = mech_df

col_list=['invalid_ballots', 'valid_ballots']
df['sum'] = df[col_list].sum(axis=1)
display(df)
```

```python
df['sum_city'] = df.groupby('reg', sort=True)["sum"].transform('sum')
display(df)
```

```python
# Формируем столбец "Плотность населения" по формуле: (Население населённого пункта / Сумму населения по региону) и округляем

df['population_density'] = (df['sum']/ df['sum_city']).round(2)
display(df)
```

```python
# Наблюдаем за отклонениями в рассчёте плотности населения, получаем удовлетворительные

df['check'] = df.groupby('reg', sort=False)["population_density"].transform('sum')
display(df)
```

```python
correlation_df=df[['population_group', 'population_density']]
correlation_df.to_csv('correlation_df.csv', index=False)

correlation_df=correlation_df.sort_values(by=['population_group'])
correlation_df=correlation_df.reset_index()
correlation_df=correlation_df.drop('index',axis=1)

display(correlation_df)
```

```python
# Самостоятельно рассчитываем коэффициент корреляции Спирмена

# Создаем столбец dx(порядок)
dx=[]
for i in range(len(correlation_df)):
    dx.append(i)
    

correlation_df['dx']=dx

# Создаем столбец dy, сортируя y 
dy=[]
y=correlation_df.population_density.tolist()
print('Столбец y:',y)
correlation_df['dy']=np.argsort(y)

# Рассчитываем столбец dd^2 (|dy-dx|^2)
correlation_df['dd2'] = (correlation_df['dx']-correlation_df['dy'])**2
display(correlation_df)
```

```python
# Рассчитываем сам коэффициент корреляции Спирмена

correlation =1-6*(correlation_df['dd2'].sum()/(len(correlation_df)**3-len(correlation_df)))
correlation
```

```python
correlation_df=correlation_df.drop([ 'dx','dy','dd2'],  axis=1)
```
